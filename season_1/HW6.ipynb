##### 1) Написать бота который будет проверять доступность сайта

```python
import telebot
from telebot import types
import requests as rq
import os
from dotenv import load_dotenv

load_dotenv()

TOKEN = os.getenv("TELEGRAM_API")
bot = telebot.TeleBot(TOKEN)

@bot.message_handler(content_types=['text'])
def start(message):
    if message.text == '/start':
        bot.send_message(message.from_user.id, "Какой сайт проверить на доступность?")
        bot.register_next_step_handler(message, get_url)
    else:
        bot.send_message(message.from_user.id, 'Напиши /start')

def get_url(message):
    try:
        rq.get(message.text)
        bot.send_message(message.from_user.id, 'Сайт работает.')
    except:
        bot.send_message(message.from_user.id, 'Сайт недоступен.')

bot.polling(none_stop=True, interval=0)
```

##### 2) Написать бота который будет анализировать текст - (Статистика, количество уникальных слов, самые популярные слова, кроме союзов и предлогов, количество предложений)

```python
import telebot
from telebot import types
import requests as rq
import os
from dotenv import load_dotenv

load_dotenv()

TOKEN = os.getenv("TELEGRAM_API")
bot = telebot.TeleBot(TOKEN)

@bot.message_handler(content_types=['text'])
def start(message):
    if message.text == '/start':
        bot.send_message(message.from_user.id, "Введите текст для статистики.")
        bot.register_next_step_handler(message, get_text)
    else:
        bot.send_message(message.from_user.id, 'Напиши /start')

def get_text(message):
    text = message.text
    text = text.lower()

    numb_symb = len(text)

    text = text.replace('!', '.')
    text = text.replace('?', '.')
    text = text.replace('...', '.')
    sents = text.split('.')

    numb_sents = len(sents)

    text = text.replace('.', '')
    text = text.replace(' - ', ' ')
    text = text.replace(',', '')

    words_dict = {}

    words = text.split(' ')

    numb_uniq_words = set(words)

    for word in words:
            if word in words_dict:
                val = words_dict.get(word) + 1
                words_dict.update({word:val})
            else:
                new_word = {word:1}
                words_dict.update(new_word)

    count = 0
    most_words = []
    for key, val in words_dict.items():
        if val > count:
            most_words.clear()
            count = val
            most_words.append(key)
        elif val == count:
            most_words.append(key)
    
    bot.send_message(message.from_user.id, f'Количество символов - {numb_symb};\nКоличество предложений - {numb_sents};\nКоличество уникальных слов - {len(numb_uniq_words)};\nСамые популярные слова: {most_words}')

bot.polling(none_stop=True, interval=0)
```
